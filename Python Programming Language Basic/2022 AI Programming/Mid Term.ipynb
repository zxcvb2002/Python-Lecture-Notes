```python
2. “+”,“-”,“*”와 변수 a와 b를 입력받아서 그 값이 아래와 같이 동시에 출력되게 코딩하여라
   결과 : 4+4=8 , 4*4=16 , 4-4=0
```

```python
a = int(input("a: "))
b = int(input("b: "))

opr = str(input("여기에 연산자를 입력하시오: "))

if opr == "+":
    print(a, opr, b, "=", a + b)
elif opr == "-":
    print(a, opr, b, "=", a - b)
elif opr == "*":
    print(a, opr, b, "=", a * b)
```

```python
8. 다음 별표를 for 문을 포함한 함수을 사용하여 출력하라
          *
         ***
        *****
         ***
          *
```

```python
# 초반 2열 / 후반 3열
for i in range(1,3):
    for j in range(4-i):
        print(" ", end="")
    for i in range(2*i-1):
        print("*", end="")
    print()
    
for i in range(1,4):
    for j in range(i):
        print(" ", end="")
    for i in range(7-2*i):
        print("*", end="")
    print()
```

```python
# 초반 3열 / 후반 2열

for i in range(1,4):
    for j in range(4-i):
        print(" ", end="")
    for i in range(2*i-1):
        print("*", end="")
    print()

for i in range(1,3):
    for j in range(i+1):
        print(" ", end="")
    for i in range(5-2*i):
        print("*", end="")
    print()
```
