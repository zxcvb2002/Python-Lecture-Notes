```python
f, axes = plt.subplots(2,2)
f.suptitle('color difference')
axes[0][0].plot(a,color = "red")
axes[0][1].plot(a*3,color = "blue")
axes[1][0].plot(a**2,color = "yellow")
axes[1][1].plot(b,color = "cyan")
axes[0][0].set_title("line")
axes[0][1].set_title("kim")
axes[1][0].set_title("curve")
axes[1][1].set_title("cyan")
f.tight_layout()
plt.show()
```
