```python
# 0. 준비

import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. 데이터

x= []
y= []
weight = 2
bios = 2
# y = 2x + 2
# -5 < x < 5
for i in range (-5,6):
  x.append(i)
  y.append(i * weight + bios)
```

```python
# 2. Hyperparameter 및 weight/bios 초기화

numOfdata = 10 # 데이터의 갯수
a = 0.01 # learning rate - 학습률 : Gradient Descent에서 얼마나 기울기를 조정할지를 정함 - Hyperparameter
numOflearning = 200 # 학습횟수
w = random.uniform(-0.1,0.1) # 맨 처음 Gradient Descent를 시행하기 전에 사용할 weight 값
b = random.uniform(-0.1,0.1) # 맨 처음 Gradient Descent를 시행하기 전에 사용할 bios 값
```

```python
# 3. 함수 설정 (MSE, 다음 weight 얻기, 다음 bios 얻기)

def get_cost(w,b,x,y): # MSE
  cost = 0
  for i in range(numOfdata):
    cost += ((w*x[i]+b-y[i])**2)
  cost = cost / numOfdata
  return cost

def get_nextweight(w,b,x,y):
  p_cost=0
  for i in range(numOfdata):
    p_cost += 2*(w*x[i]+b-y[i])*x[i]
  weight = w - ((a*p_cost) / numOfdata)
  return weight

def get_nextb(w,b,x,y):
  b_cost = 0
  for i in range(numOfdata):
    b_cost += 2*(w*x[i]+b-y[i])
  bios = b-a*b_cost/numOfdata
  return bios
```

```python
# 4. 학습시키기

W_val = []
b_val = []
cost_val = []
num = []

for X in range(numOflearning):
  w = get_nextweight(w,b,x,y)
  b = get_nextb(w,b,x,y)
  c = get_cost(w,b,x,y)
  W_val.append(w)
  b_val.append(b)
  print("W =", w, end=" ")
  cost_val.append(c)
  print("cost =", c)
  num.append(X)
  print()
```

```python
# 5. 최적회된 값 출력, 그래프로 cost값의 변화과정 보기

plt.plot(num,cost_val)
plt.show()
print("Optimal w =", w, "b =", b, "cost =", c)
```
