```python
x = [30, 100, 120, 170, 50]
y = [20, 60, 70, 80, 25]

iteration = 100
w_n = 0.53
a = 0.0001
m = len(x)
# loss = 1/m*sig(w*x-y)**2

for j in range(1,iteration+1):
    sigma = 0
    for i in range(m):
        sigma += ((w_n * x[i] - y[i]) * x[i]) 
    loss_m = sigma * 2 / m

    for i in range(m):
        sigma += ((w_n*x[i])-y[i])**2
    
    loss = sigma/m

    w_n1 = w_n - (a * loss_m)

    print('iteration =', j, '가중치 =',w_n1, 'Loss =',loss)
    w_n = w_n1
```
